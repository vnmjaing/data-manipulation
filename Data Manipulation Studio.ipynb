## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

baltimore = pd.read_csv(r"dataset\baltimore_9-24-2016_9-30-2017.csv")
boston = pd.read_csv(r"dataset\boston_9-24-2016_9-30-2017.csv")
new_york = pd.read_csv(r"dataset\new-york_9-24-2016_9-30-2017.csv")
philly = pd.read_csv(r"dataset\philadelphia_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
for col in baltimore.columns:
    pct_missing = np.mean(baltimore[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
balt_pct_missing = baltimore.isna().mean()
balt_pct_missing[balt_pct_missing >=.9]*100

balt_less_missing_cols = baltimore.drop(columns=balt_pct_missing[balt_pct_missing >=.9].index)
balt_less_missing_cols.shape
```

```python
# Clean your data here!
for col in boston.columns:
    pct_missing = np.mean(boston[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
boston_pct_missing = boston.isna().mean()
boston_pct_missing[boston_pct_missing >=.9]*100

boston_less_missing_cols = boston.drop(columns=boston_pct_missing[boston_pct_missing >=.9].index)
boston_less_missing_cols.shape
```

```python
# Clean your data here!
for col in new_york.columns:
    pct_missing = np.mean(new_york[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
new_york_pct_missing = new_york.isna().mean()
new_york_pct_missing[new_york_pct_missing >=.90]*100

new_york_less_missing_cols = new_york.drop(columns=new_york_pct_missing[new_york_pct_missing >=.9].index)
new_york_less_missing_cols.shape
```

```python
# Clean your data here!
for col in philly.columns:
    pct_missing = np.mean(philly[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
philly_pct_missing = philly.isna().mean()
philly_pct_missing[philly_pct_missing >=.90]*100

philly_less_missing_cols = philly.drop(columns=philly_pct_missing[philly_pct_missing >=.9].index)
philly_less_missing_cols.shape[0]
```

```python
print(philly_less_missing_cols.shape[0] + new_york_less_missing_cols.shape[0] + boston_less_missing_cols.shape[0] + balt_less_missing_cols.shape[0]) 
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
northeast_pumpkins = pd.concat([balt_less_missing_cols, boston_less_missing_cols, new_york_less_missing_cols, philly_less_missing_cols
                                                   ], ignore_index = True)
northeast_pumpkins
```
